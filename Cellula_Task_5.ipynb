## Data Preprocessing

```python
# Fill missing values for categorical columns with mode
categorical_columns = ['Car Condition', 'Weather', 'Traffic Condition']
for column in categorical_columns:
    df[column] = df[column].fillna(df[column].mode()[0])
```

```python
# Fill missing values for numerical columns with mean
numerical_columns = ['fare_amount', 'pickup_longitude', 'pickup_latitude', 'dropoff_longitude', 'dropoff_latitude', 'passenger_count', 'hour', 'day', 'month', 'weekday', 'year', 'jfk_dist', 'ewr_dist', 'lga_dist', 'sol_dist', 'nyc_dist', 'distance', 'bearing']
for column in numerical_columns:
    df[column] = df[column].fillna(df[column].mean())
```

```python
# List of numerical columns to check for outliers
numerical_columns = ['fare_amount', 'pickup_longitude', 'pickup_latitude', 
                     'dropoff_longitude', 'dropoff_latitude', 'passenger_count', 
                     'hour', 'day', 'month', 'weekday', 'year', 
                     'jfk_dist', 'ewr_dist', 'lga_dist', 'sol_dist', 
                     'nyc_dist', 'distance', 'bearing']

# Define a function to handle outliers using the IQR method
def handle_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
    return df

# Apply the outlier handling function to each numerical column
for column in numerical_columns:
    df = handle_outliers_iqr(df, column)

```
